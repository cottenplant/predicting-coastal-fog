```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
plt.rcParams['text.color'] = 'k'
plt.rcParams['figure.figsize'] = 12, 8
```

```python
data = pd.read_csv("data/NOAA_LAX_Data.csv")
```

```python
def null_precip(col):
    if col == 99.99:
        return 0.0
    else:
        return col

df = data.loc[data['STN---'] == 722950].copy()
df.reset_index(inplace=True)
df.columns = df.columns.str.strip()
df.drop(['Unnamed: 0', 'index', 'STN---', 'WBAN', '', '.1', '.2', '.3', '.4', '.5', 
             'Unnamed: 22'], axis=1, inplace=True)
df['YEARMODA'] = pd.to_datetime(df['YEARMODA'], format='%Y%m%d')
df['month'] = df['YEARMODA'].apply(lambda date: date.month)
df['year'] = df['YEARMODA'].apply(lambda date: date.year)
df['MAX'] = df['MAX'].apply(lambda col: float(col.strip('*')))
df['MIN'] = df['MIN'].apply(lambda col: float(col.strip('*')))
df['PRCP'] = df['PRCP'].apply(lambda col: float(col[:-1].strip()))
df['PRCP'] = df['PRCP'].apply(null_precip)
df['dummy'] = 1
```
