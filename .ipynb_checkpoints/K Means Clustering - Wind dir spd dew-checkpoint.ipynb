```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import pandas as pd
from math import radians
```

```python
data_file = "data/wind_speed_dewpt.csv"
```

```python
# Read df into pandas
df = pd.read_csv(data_file)
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df.columns = ['dir', 'spd', 'dew']
```

```python
df.head()
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(8,6))

sns.kdeplot(df['meanwdird'], shade=True)

plt.title("Distribution of Mean Wind Bearing vs Wind Speed", fontsize=18);
```

```python
plt.figure(figsize=(8,8))

ax = plt.subplot(111, polar=True)
ax.scatter(x=[radians(x) for x in df['meanwdird'].values], 
           y=df['meanwindspdm'], alpha=0.3)
ax.set_theta_zero_location('N')
ax.set_theta_direction(-1)

plt.title('Mean Wind Bearing (°) vs Mean Windspeed (km/h)', fontsize=18);
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(8,8))

ax = plt.subplot(111, polar=True)
ax.scatter(x=[radians(x) for x in df['meanwdird'].values], 
           y=df['meandewptm'], alpha=0.3)
ax.set_theta_zero_location('N')
ax.set_theta_direction(-1)

plt.title('Mean Wind Dir (°) vs Mean Dew Temp', fontsize=18);
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), cmap='coolwarm');
```

```python
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['dir'], df['spd'], df['dew'], c='skyblue', s=25)
plt.xlabel('wind dir degrees')
plt.ylabel('wind spd kmh')
plt.clabel('dewpoint degrees C')
ax.view_init(25, 120)
plt.show()
```

```python
x, y, c = ['dir', 'spd', 'dew']
```

```python
x_label = df[x]
y_label = df[y]
c_label = df[c]
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(df[[x, y, c]])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
fig, (ax1,ax2) = plt.subplots(1, 2, sharey=True, figsize=(12,10))

ax1.set_title('Original')
ax1 = fig.add_subplot(111, projection='3d')
ax1.scatter(x_label, y_label, c=c_label, cmap='coolwarm')

ax2.set_title('KMeans')
ax2 = fig.add_subplot(111, projection='3d')
ax2.scatter(x_label, y_label, c=c_label, cmap='rainbow')

ax1.view_init(25, 120)
ax2.view_init(35, 135)
```
