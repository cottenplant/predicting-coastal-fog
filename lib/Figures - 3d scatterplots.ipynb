```python
import clean_data
df_smo = clean_data.init()

import numpy as np
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

import seaborn as sns
sns.set_style('white')

plt.rcParams['figure.figsize'] = 12, 8

fig_dir = r'figures/'
```

# Rain

```python
graphs = {'no rain': df_smo[df_smo['rain'] == 0], 
          'rain': df_smo[df_smo['rain'] == 1], 
          'rain merged': df_smo}

colors = {'no rain': 'coolwarm',
          'rain': 'coolwarm_r', 
          'rain merged': 'coolwarm'}

for name, df in graphs.items():
    fig = plt.figure(figsize=(12,8))
    ax = fig.add_subplot(111, projection='3d')
    
    ax.scatter(df['mdir'], df['mspd'], df['mdew'], 
               c=df['rain'], cmap=colors[name], s=30)

    ax.view_init(25, 120)
 
    ax.set_xlim(0, 360)
    ax.set_ylim(0, 50)
    ax.set_zlim(-30, 30)

    ax.set_xlabel('wind dir (daily)')
    ax.set_ylabel('wind spd (daily)')
    ax.set_zlabel('dew pt (daily)')

    plt.title(name, fontsize=20)
    plt.savefig(fig_dir + name + '_3d.png', dpi=300)
```
