```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# plt.rcParams['figure.figsize'] = (12, 8)
```

```python
import june_gloom
june_gloom.init()
june_gloom.noaa_clean()
june_gloom.feature_sst_anomaly()
from june_gloom import df_lax
```

```python
# Copy df's for ease
df_lax_fog = df_lax[df_lax['FRSHTT']==100000]
df_lax_rain = df_lax[df_lax['FRSHTT']==10000]
df_lax_sun = df_lax[df_lax['FRSHTT']==0]
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(12,8))

sns.kdeplot(df_lax_fog['sst_anomaly'].dropna(), color='grey', shade=True)
sns.kdeplot(df_lax_rain['sst_anomaly'].dropna(), color='blue', shade=True)
sns.kdeplot(df_lax_sun['sst_anomaly'].dropna(), color='red', shade=True)

plt.legend(['Fog', 'Rain', 'Sun']);
```

```python
df_lax_rain[df_lax_rain['PRCP']>1.0]
```

```python
with sns.axes_style('white'):
    sns.jointplot('PRCP', 'sst_anomaly', data=df_lax_rain[df_lax_rain['PRCP']>0.1], 
                  kind='kde', size=10);
```

```python
# Must use 3 month running averages to compare with SST data (also 3 month running avgs)

with sns.axes_style('whitegrid'):
    sns.jointplot('MAX', 'sst_anomaly', data=df_lax_fog, kind='reg', size=10, 
                  color='red');
```

```python
with sns.axes_style('white'):
    sns.jointplot('MAX', 'sst_anomaly', data=df_lax_fog, kind='scatter', size=10);
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(12,8))

sns.distplot(df_lax_fog['sst_anomaly'].dropna(), bins=40)
plt.title("Distribution of SST Anomaly on 4,249 Foggy Days (no rain) from 1973-2018", 
          fontsize=20)

plt.tight_layout()
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(12,8))

sns.distplot(df_lax_rain['sst_anomaly'].dropna(), bins=40)
plt.title("Distribution of SST Anomaly on 1,169 Rainy Days from 1973-2018", fontsize=20)

plt.tight_layout()
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(12,8))

sns.distplot(df_lax_sun['sst_anomaly'].dropna(), bins=40)
plt.title("Distribution of SST Anomaly on 10,009 Sunny Days from 1973-2018", fontsize=20)

plt.tight_layout()
```
